```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

Check if there are any null values

```python
df.isnull().sum()
```

Fill null values with their mean (Revenue and Metascore columns)

```python
df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean(), inplace=True)
df['Metascore'].fillna(df['Metascore'].mean(), inplace=True)
```

```python
df.info()
```

# Which genre has the most movies? (top 10)

```python
df['Genre'].str.split(',').explode().value_counts(normalize=True).head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 movie genres')
#Drama has the highest number of movies
```

# Which year has the highest number of released movies?

```python
df['Year'].value_counts().sort_index().plot()
plt.xticks(range(2006,2017))
plt.xlabel('Year')
plt.ylabel('Number of movies')
plt.show()
#The number of released movies increases each year and it reached its peak in 2016
```

# What are the top 10 highest grossing movies?

```python
x = df.sort_values(by=['Revenue (Millions)'], ascending=False, inplace=False).Title.head(10)
y= df.sort_values(by=['Revenue (Millions)'], ascending=False, inplace=False)['Revenue (Millions)'].head(10)
fig, ax = plt.subplots()
ax.barh(x, y)
ax.invert_yaxis()
```

# Does the runtime have an impact on the rating?

```python
df.plot(x="Rating",y="Runtime (Minutes)",kind="scatter")
plt.xlabel("Rating")
plt.ylabel("Runtime (Minutes)")
plt.title("Rating vs Runtime")
plt.show()
# viewers tend to love movie that last 80-140 minutes.
```

# Does the metascore have an impact on the rating?

```python
df.plot(x="Rating", y="Metascore", kind="scatter")
plt.xlabel("Rating")
plt.ylabel("Metascore")
plt.title("Rating vs Metascore")
plt.show()
#The higher the Metascore, the higher the rating.
```
